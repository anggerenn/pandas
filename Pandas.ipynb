# Making changes to the data

```python
# Make a Total column by Add all the stats
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']

df.head()
```

```python
# Delete Total column
df = df.drop(columns=['Total'])

# Make a Total column by .sum()
df['Total'] = df.iloc[:, 4:10].sum(axis=1)

# Rearrange Total column beside Speed column
df = df[['#', 'Name', 'Type 1', 'Type 2', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Total', 'Generation', 'Legendary']]
df.head()
```

```python
df = pd.read_csv('pokemon_data.csv')

df['Total'] = df.iloc[:, 4:10].sum(axis=1)

cols = list(df.columns)
df = df[cols[0:10] + [cols[-1]] + cols[10:12]]

df.head()
```

# Filtering Data

```python
# Filtered by Type 1 Grass, Type 2 Posion, and HP > 60
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

#new_df.to_csv('filtered.csv', index=False)

new_df
```

```python
# Reset index
new_df.reset_index(drop=True, inplace=True)

new_df
```

```python
# Filtered out "Mega" pokemon in df
nomega_df = df.loc[~df['Name'].str.contains('Mega')]

nomega_df
```

```python
# Filter using regex
import re

df.loc[df['Type 1'].str.contains('fire|poison', flags=re.I , regex=True)]
```

```python
# Get Pokemon Name starts with 'Pi' and Type 1 Electric
df.loc[(df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)) & df['Type 1'].str.contains('electric', flags=re.I, regex=True)]
```

# Conditional changes

```python
df.loc[df['Type 1'] == "Fire", 'Legendary'] = True

df.loc[df['Type 1'] == "Fire"].head()
```

# Aggregate Statistics (Groupby)

```python
# Use modified.csv 
df = pd.read_csv('modified.csv')

df.groupby(['Type 1', 'Type 2']).mean().sort_values('HP', ascending=False)
```

```python
df = pd.read_csv('modified.csv')

df['Count'] = 1

df.groupby(['Type 1', 'Type 2']).count()['Count']
```
